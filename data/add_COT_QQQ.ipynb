```python
import pandas as pd
import numpy as np
import zipfile, urllib.request, shutil
import os
import matplotlib.pyplot as plt
```

```python
#Load COT data
# Data source: https://www.cftc.gov/MarketReports/CommitmentsofTraders/index.htm

data_2020 = pd.read_excel('2020.xls')
data_2019= pd.read_excel('2019.xls')
data_2018=pd.read_excel('2018.xls')
data_2017=pd.read_excel('2017.xls')
```

```python
data_2020
```

```python
data_2020 = data_2020[['Market_and_Exchange_Names', 
                       'Report_Date_as_MM_DD_YYYY',
                       'Open_Interest_All',
                       'Tot_Rept_Positions_Long_All',
                       'Tot_Rept_Positions_Short_All',
                       'NonRept_Positions_Long_All',
                       'NonRept_Positions_Short_All',]]

data_2019 = data_2019[['Market_and_Exchange_Names', 
                       'Report_Date_as_MM_DD_YYYY',
                       'Open_Interest_All',
                       'Tot_Rept_Positions_Long_All',
                       'Tot_Rept_Positions_Short_All',
                       'NonRept_Positions_Long_All',
                       'NonRept_Positions_Short_All',]]

data_2018 = data_2018[['Market_and_Exchange_Names', 
                       'Report_Date_as_MM_DD_YYYY',
                       'Open_Interest_All',
                       'Tot_Rept_Positions_Long_All',
                       'Tot_Rept_Positions_Short_All',
                       'NonRept_Positions_Long_All',
                       'NonRept_Positions_Short_All',]]

data_2017 = data_2017[['Market_and_Exchange_Names', 
                       'Report_Date_as_MM_DD_YYYY',
                       'Open_Interest_All',
                       'Tot_Rept_Positions_Long_All',
                       'Tot_Rept_Positions_Short_All',
                       'NonRept_Positions_Long_All',
                       'NonRept_Positions_Short_All',]]
```

```python
frames = [data_2017, data_2018, data_2019, data_2020]
COT=pd.concat(frames)
```

```python
COT=COT[COT['Market_and_Exchange_Names'].str.contains("BITCOIN")].copy()
```

```python
COT['Date']=pd.to_datetime(COT['Report_Date_as_MM_DD_YYYY'],format='%m%d%y')
#sort by date
COT.sort_values(by=['Date'],inplace=True)
COT.set_index('Date', inplace=True)
```

```python
#remove unnesssary column
COT.drop(['Market_and_Exchange_Names', 'Report_Date_as_MM_DD_YYYY'], axis=1, inplace=True)
```

```python
COT
```

```python
QQQ_prices = pd.read_csv('QQQ_FINX_RIOT.csv',
                     date_parser=lambda dt: pd.to_datetime(dt, format='%Y-%m-%d'),
                     index_col = 1).dropna()

QQQ_prices.drop(["PERMNO"], axis=1, inplace=True)
QQQ_prices = QQQ_prices.pivot_table('PRC', index='date',columns= 'TICKER')
QQQ_prices.head()
```

```python
# load the new file
dataset = pd.read_csv('btc_dataset.csv', header=0, infer_datetime_format=True, parse_dates=['Date'], index_col=['Date']) 

dataset = dataset.astype(float)
df=dataset['2017-12-13':'2020-10-06']
```

```python
df_QQQ=df.join(QQQ_prices)
df_QQQ
```

```python
df_COT_QQQ=df_QQQ.join(COT)
df_COT_QQQ
```

```python
df_COT_QQQ=df_COT_QQQ.ffill()
df_COT_QQQ
```

```python
df_COT_QQQ=df_COT_QQQ.bfill()
df_COT_QQQ
```

```python
df_COT_QQQ.to_csv('btc_dataset_COTQQQ.csv')
```
